## shutil

* .copy() - копировать файл
* .copytree() - копировать папку
* .move() - перемещает и файлы и папки
* .rmtree() - удаляет папку
* .make_archive(<имя_архива_без_расширения>, <формат_архива(zip)>, <путь>) 

```python
import shutil
```

```python
shutil.copy('ass_we_can.txt', 'ass_we_can_copy.txt')
```

```python
def func(stop=False):
    if stop:
        return 0
    return func()
```

```python
def loop(a):
    if not a:
        return
    print(a[0])
    loop(a[1:])
    
    loop([1, 2, 3])
```

```python
loop([1, 2, 3])
```

```python
def countdown(x):
    if x == 0: return
    else: 
        print(x)
        countdown(x-1)
```

```python
countdown(10)
```

```python
def factorial(n):
    if n <= 1:
        return 1
    else:
        return n * factorial(n-1)
        
```

```python
factorial(3)
```

```python
# fibonachi
# 4 = 1 + 2 + 3
def fib(n):
    if n <= 1:
        return n
    else:
        return fib(n-1) + fib(n-2)
fib(4)
```
