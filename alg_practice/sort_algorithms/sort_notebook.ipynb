# Bubble Sort

```python
def bubble_sort(arr):
    size = len(arr)
    swapped = False
    
    while size != 0 or not swapped:
        for i in range(size-1):
            if arr[i] > arr[i+1]: 
                arr[i], arr[i+1] = arr[i+1], arr[i]
                swapped = True
         
        if not swapped: break     
        size -= 1
    
    return arr
```

```python
sarr = [4, 2, 6]
elems = [5, 9, 2, 1, 67, 34, 88, 34]
bubble_sort(elems)
```

# Insertion Sort

```python
def insertion_sort(arr):
    for i in range(1, len(arr)):
        anchor = arr[i]
        j = i - 1
        
        while j >= 0 and anchor < arr[j]:
            arr[j+1] = arr[j]
            j -= 1
        arr[j+1] = anchor
    return arr
```

```python
elems = [5, 9, 2, 1, 67, 34, 88, 34]
insertion_sort(elems)
```

```python
# implementation of quick sort in python using hoare partition scheme

def swap(a, b, arr):
    if a!=b:
        tmp = arr[a]
        arr[a] = arr[b]
        arr[b] = tmp

def quick_sort(elements, start, end):
    if start < end:
        pi = partition(elements, start, end)
        quick_sort(elements, start, pi-1)
        quick_sort(elements, pi+1, end)

def partition(elements, start, end):
    pivot_index = start
    pivot = elements[pivot_index]

    while start < end:
        while start < len(elements) and elements[start] <= pivot:
            start+=1

        while elements[end] > pivot:
            end-=1

        if start < end:
            swap(start, end, elements)

    # swap(pivot_index, end, elements)
    elements[pivot_index], elements[end] = elements[end], elements[pivot_index]

    return end
```

## Shellsort

```python
def shellsort(arr):
    size = len(arr)
    gap = size // 2
    
    while gap > 0:
        for i in range(gap, size):
            anchor = arr[i]
            j = i
            while j >= gap and arr[j - gap] > anchor:
                arr[j] = arr[j - gap]
                j -= gap
            arr[j] = anchor
        
        gap //= 2
            
```

# Selection sort

```python
def find_min(arr):
    min = arr[0]
    idx = 0
    for i in range(len(arr)):
        if arr[i] < min: 
            min = arr[i]
            idx = i
    return min, idx

def selection_sort(arr):
    size = len(arr)
    for i in range(size-1):
        min_i = i
        for j in range(min_i+1, size):
            if arr[j] < arr[min_i]: min_i = j

        if i != min_i: arr[i], arr[min_i] = arr[min_i], arr[i]
```

```python
elements = [78, 12, 15, 8, 61, 53, 23, 27]
print(elements)
selection_sort(elements)
elements
```
