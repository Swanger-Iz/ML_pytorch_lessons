### Dataset

```python
from torch.utils.data import Dataset
```

```python
class JointDS(Dataset):
    def __init__(self, x, y):
        self.x = x
        self.y = y
        
    def __len__(self):
        return len(self.x)
    
    def __getitem__(self, index):
        return self.x[index], self.y[index]
```

```python
from torch.utils.data import TensorDataset
jDS = TensorDataset(t_x, t_y)
```

```python
for i in jDS:
    print(i)
```

```python
it = iter(jDS)
next(it)
```

```python
dataloader = DataLoader(jDS, batch_size=2, shuffle=True)
for i, batch in enumerate(dataloader):
    print(f'Batch {i}: | X: {batch[0]} | y: {batch[1]}')
```

```python
for epoch in range(2):
    print('Epoch', epoch+1)
    
    for i, batch in enumerate(dataloader, 1):
        print(f'Batch {i}: | X: {batch[0]} | y: {batch[1]}')
```

```python
from torch.utils.data import DataLoader, Dataset 
```

```python
class PetsDS(Dataset):
    def __init__(self, file_list, labels):
        self.file_list = file_list
        self.labels = labels
        
    def __getitem__(self, index):
        file = self.file_list[index]
        label = self.labels[index]
        return file, label
    
    def __len__(self):
        return len(self.labels)
        
```

```python
import torchvision.transforms as transforms
```

```python
img_height, image_width = 80, 120
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Resize((img_height, image_width))
])
```

```python
class PetsDS(Dataset):
    def __init__(self, file_list, labels, transform=None):
        self.file_list = file_list
        self.labels = labels
        self.transform = transform
        
    def __getitem__(self, index):
        img = Image.open(self.file_list[index])
        if self.transform is not None: img = self.transform(img)
        label = self.labels[index]
        return img, label
    
    def __len__(self):
        return len(self.labels)
        
```

```python
pets_tds = PetsDS(all_files, labels, transform)
```

```python
counter = 12499 - 3
t_ = []
while counter < 12499 + 3:
    t_.append(pets_tds[counter])
    counter += 1
len(t_)
```

```python
fig = plt.figure(figsize=[10, 6])
for i, ex in enumerate(t_):
    ax = fig.add_subplot(2, 3, i+1)
    ax.set_xticks([]); ax.set_yticks([])
    ax.imshow(ex[0].numpy().transpose((1, 2, 0)))
    ax.set_title(f'{ex[1]}', size=15)
plt.tight_layout()
plt.show()
```
