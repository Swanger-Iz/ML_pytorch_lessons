## Creating DB, and import libs

```python
from sqlalchemy import create_engine, Column, Integer, String
from sqlalchemy.orm import declarative_base
```

```python
db_url = "sqlite:///database_nt.db"
engine = create_engine(db_url)
Base = declarative_base()
```

```python
class User(Base):
    __tablename__ = 'users'
    
    id = Column(Integer, primary_key=True)
    name = Column(String)
    age = Column(Integer)

class Gachi(Base):
    __tablename__ = "gachi_actors"

    id = Column(Integer, primary_key=True)
    name = Column(String)
    hair_color = Column(String)
    phrase = Column(String)

Base.metadata.create_all(engine)
```

```python
from sqlalchemy.orm import sessionmaker
Session = sessionmaker(bind=engine)
session = Session()
```

```python
### Inserting rows
user = User(name="John doe", age=30)
user2 = User(name="Andrew Pip", age=25)
user3 = User(name="Iron man", age=57)
user4 = User(name="Richard Gear", age=25)

session.add(user2)
session.add_all([user3, user4])
session.commit()
```

## Filter

```python
# Filter .filter | filter_by, (.one(), one_or_none(), all(), .first())
user = session.query(User).filter_by(id=1).one()
user.id, user.name, user.age
```

```python
users = session.query(User).all()
[(u.id, u.name, u.age) for u in users]
```

```python
# Update
user = session.query(User).filter_by(id=1).one()
print(user.id, user.name, user.age)

user.name = "Another Name"
user.age = 88

session.commit()

print(user.id, user.name, user.age)
```

```python
users = session.query(User).all()
[(u.id, u.name, u.age) for u in users]
```

## Delete

```python
# Delete
user = session.query(User).filter(User.name == 'Another Name').one_or_none()
print(user.id, user.name, user.age)

session.delete(user)

users = session.query(User).all()
[(u.id, u.name, u.age) for u in users]
```

```python
users = session.query(User).all()
[(u.id, u.name, u.age) for u in users]
```

## Ordering

```python
import random
```

```python
names = ['Andrew Pip', 'Sara Poetry', 'Kim uv', 'Simon Pinchevski']
ages = [20, 21, 22, 23, 24, 25, 26, 30, 35, 67]

for x in range(20):
    user = User(name=random.choice(names), age=random.choice(ages))
    session.add(user)

session.commit()
```

```python
q = session.query(User).order_by(User.age).all()  # ascending
[(r.id, r.name, r.age) for r in q]
```

```python
q = session.query(User).order_by(User.age.desc()).all()  # descending
[(r.id, r.name, r.age) for r in q]
```

```python
q = session.query(User).order_by(User.age.desc(), User.name).all()  # multiple
[(r.id, r.age, r.name, ) for r in q]
```
