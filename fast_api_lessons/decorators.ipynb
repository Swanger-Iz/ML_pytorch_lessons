```python
from typing import Callable
import time
```

```python
from functools import wraps 
```

```python
def deco(func: Callable):
    def wrapper(*args, **kwargs):
        res = func(*args, **kwargs)
        return res
    return wrapper
```

```python
@deco
def get_square(num):
    return num ** 2

## OR

get_square = deco(get_square)
```

```python
def limit_calls(limit: int):
    def wrapper(func: Callable):
        @wraps(func)
        def inner(*args, **kwargs):
            nonlocal limit
            if limit == 0:
                print('Cannot to call the function')
                return
            res = func(*args, **kwargs)
            limit -= 1
            return res
        return inner
    return wrapper

@limit_calls(2)
def my_f(sleep_time: int):
    """_summary_

    Args:
        sleep_time (int): _description_

    Returns:
        _type_: _description_
    """
    time.sleep(sleep_time)
    return 1

## OR

my_f = limit_calls(2)(my_f)
```
