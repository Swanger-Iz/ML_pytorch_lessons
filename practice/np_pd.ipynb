## Код для номализации 2 мерного массива

* нужно преобразовать нормально, а потом обратно
* то есть min el = 0, max el = 1
* сделать для каждого столбца 

```python
def norm(x):
    return (x - x.min()) / (x.max() - x.min()), x.max(), x.min()
```

```python
def retrun_arr(x_norm, old_max, old_min):
    res = x_norm * (old_max - old_min) + old_min
    return res
```

```python
rand_digits
```

```python
normalized, max_old, min_old = norm(rand_digits)
normalized, max_old, min_old
```

```python
retrun_arr(normalized, max_old, min_old)
```

```python
def my_round(x):
    # x = x * 1000
    if x % 500 == 0: return x
    else:
        if int(str(x)[-3]) >= 5: return x + x % 500
        else: return x - x % 500
```
